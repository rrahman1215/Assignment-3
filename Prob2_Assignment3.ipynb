```python
%load_ext tensorboard

from datetime import datetime
from packaging import version

import tensorflow as tf
from tensorflow import keras

import tensorboard
import matplotlib.pyplot as plt

import numpy as np

(train_images, train_labels), (test_images, test_labels) = keras.datasets.cifar10.load_data()
train_images = train_images / 255.0
```

```python
x_train = train_images.reshape(-1, 32, 32, 3) #add an additional dimension to represent the single-channel
x_test = test_images.reshape(-1, 32, 32, 3)

cnn_model = tf.keras.models.Sequential()
cnn_model.add(tf.keras.layers.Conv2D(6, (5,5), padding='same', activation='relu'))     #layer 1
cnn_model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2)))           #layer 2

cnn_model.add(tf.keras.layers.Conv2D(16, (5,5), padding='same', activation='relu'))    #layer 3
cnn_model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2,2)))          #layer 4

cnn_model.add(tf.keras.layers.Conv2D(120, (5,5), padding='same', activation='relu'))   #layer 5

cnn_model.add(tf.keras.layers.Flatten())                                       
cnn_model.add(tf.keras.layers.Dense(84))                                            #layer 6
cnn_model.add(tf.keras.layers.Activation('relu'))

cnn_model.add(tf.keras.layers.Dense(10))
cnn_model.add(tf.keras.layers.Activation('softmax'))

```

```python
############################# Hyperparameters  #################################
batch_size = 32
epochs = 25
choose_learning_rate = 0.005
optim = keras.optimizers.Adam(learning_rate=choose_learning_rate)      #adam
optimizer_name = 'Adam'
#optim = tf.keras.optimizers.RMSprop(learning_rate=choose_learning_rate)  #RMS prop
#optimizer_name = 'RMSprop'
################################################################################
      
      
cnn_model.compile(loss='sparse_categorical_crossentropy', optimizer=optim, metrics=['accuracy'])
cnn_model.build(input_shape=(1,32,32,3))
cnn_model.summary()

print(optimizer_name, batch_size, choose_learning_rate)
logdir="logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir)

model1 = cnn_model
# Train the model.
model1.fit(
          x_train,
          train_labels, 
          batch_size=batch_size,
          epochs=epochs, 
          callbacks=[tensorboard_callback])

print('Evaluate with test data:')
score = model1.evaluate(x_test, test_labels)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print('\n')
        
```

```python
####Implementation of Equivalent feed forward NN #########


x_train = train_images.reshape(-1, 32, 32, 3) #add an additional dimension to represent the single-channel
x_test = test_images.reshape(-1, 32, 32, 3)

nn_model = tf.keras.models.Sequential()

nn_model.add(tf.keras.layers.Flatten())

nn_model.add(tf.keras.layers.Dense(6))             # Hidden layer 1                          
nn_model.add(tf.keras.layers.Activation('relu'))


nn_model.add(tf.keras.layers.Dense(16))             # Hidden layer 2                          
nn_model.add(tf.keras.layers.Activation('relu'))


nn_model.add(tf.keras.layers.Dense(120))             # Hidden layer 3                       
nn_model.add(tf.keras.layers.Activation('relu'))


nn_model.add(tf.keras.layers.Dense(10))            # output layer
nn_model.add(tf.keras.layers.Activation('softmax'))



################# Hyper parameters #############
batch_size = 32
epochs = 25
choose_learning_rate = 0.005
optim = keras.optimizers.Adam(learning_rate=choose_learning_rate)      #adam
################################################


nn_model.compile(loss='sparse_categorical_crossentropy', optimizer=optim, metrics=['accuracy'])
nn_model.build(input_shape=(1,32,32,3))
nn_model.summary()


logdir="logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir)

model1 = nn_model
# Train the model.
model1.fit(
    x_train,
    train_labels, 
    batch_size=batch_size,
    epochs=epochs, 
    callbacks=[tensorboard_callback])

print('Evaluate with test data:')
score = model1.evaluate(x_test, test_labels)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
